```python
import torch

from models import ManhattenDistance, EuclideanDistance, CosineSimilarity
```

```python
x1 = torch.rand((32, 1000))
x2 = torch.rand((32, 1000))
```

### Manhatten Distance

```python
manhatten = ManhattenDistance()
print(manhatten(x1, x2).shape)
```

### Euklidische Distanz

```python
euclidean = EuclideanDistance()
print(euclidean(x1, x2).shape)
```

### Cosine Similarity

```python
cosine = CosineSimilarity()
epsilon = torch.ones((32)) * 1e-8
d1 = cosine(x1, x2)
d2 = torch.div(torch.sum(torch.mul(x1, x2), dim=1), torch.mul(torch.max(torch.linalg.norm(x1, dim=1, ord=2), epsilon), torch.max(torch.linalg.norm(x2, dim=1, ord=2), epsilon)))
print(d1.shape)
print(d2.shape)
print(torch.isclose(d1, d2))
```
